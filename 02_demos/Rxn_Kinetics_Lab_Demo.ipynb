# Rxn Kinetics Demo

## Download Packages

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Dataset

```python
t = np.arange(0,40,5)
print(t)
C = np.array([20, 10.5, 7, 5, 4, 3.5, 3, 2.5])
print(C)
```

## Zero Order?

```python
plt.plot(t,C,'*')
plt.xlabel('time [min]')
plt.ylabel('Concentration [mg/L]')
slope, intercept, r_value, p_value, std_err = stats.linregress(t,C)
fit = slope*t + intercept
plt.plot(t,fit,'k-')
plt.title('Zero Order?: R^2 = ' + str(r_value**2))
```

## First Order?

```python
plt.plot(t,np.log(C),'*')
plt.xlabel('time [min]')
plt.ylabel('ln(C)')
slope, intercept, r_value, p_value, std_err = stats.linregress(t,np.log(C))
fit = slope*t + intercept
plt.plot(t,fit,'k-')
plt.title('First Order?: R^2 = ' + str(r_value**2))
```

## Second Order?

```python
plt.plot(t,1/C,'*')
plt.xlabel('time [min]')
plt.ylabel('1/C [L/mg]')
slope, intercept, r_value, p_value, std_err = stats.linregress(t,1/C)
fit = slope*t + intercept
plt.plot(t,fit,'k-')
plt.title('Second Order?: R^2 = ' + str(r_value**2))
```
